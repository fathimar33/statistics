can check the association between the other columns with price using chi square test.If not associated it can be removed

```python
from  scipy.stats import chi2_contingency
```

```python
num_columns.dtypes
```

```python
#Checking the car's ID and symboling is unnecessary as it is not meaningful.
```

The chi-square test is typically employed for discrete random variables. Therefore, excluding columns with float datatype, as there are alternative methods available to handle them.

```python
from scipy.stats import chi2
for column in num_columns.columns:
    if num_columns[column].dtype == "int64":
        if column == "car_ID" or column == "symboling":
            continue
        else:
            l = []
            l.append([num_columns[column].values, num_columns["price"].values])
            dataf=pd.DataFrame(l)
            chi2statistic, pvalue, _, _ = chi2_contingency(l) #since degrees of freedom and expected frequency are not necessarty to display I used underscore
            dof=(dataf.shape[0]-1)*(dataf.shape[1]-1)
            alpha=0.05
            if chi2statistic > chi2.ppf(1-alpha,dof):
                print(column, "vs price: association")
            else:
                print(column, "vs price: no association")
```
