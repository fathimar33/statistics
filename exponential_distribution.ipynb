```python
import numpy as np
mean = 20000
rate_parameter = 1 / mean
dataset_size = 1000
exponential_data = np.random.exponential(scale=1/rate_parameter, size=dataset_size)
print(exponential_data)
```

```python
import pandas as pd
df=pd.DataFrame(exponential_data)
```

```python
df
```

```python
df.rename(columns={0:"profit"},inplace=True)
```

```python
import matplotlib.pyplot as plt
profits = df["profit"]
days = list(range(1, len(profits) + 1))
# Plot the data on a logarithmic scale
plt.scatter(days, profits)
plt.yscale('log')  # Set the y-axis scale to logarithmic
plt.xlabel('Day')
plt.ylabel('Profit')
plt.title('Profit Trend (Logarithmic Scale)')
plt.show()
```

```python
import seaborn as sns
sns.histplot(df["profit"],kde=True,bins=14)
```

```python
mean=df["profit"].mean()
```

```python
mean
```

```python
scale=1/mean
```

```python
from scipy.stats import expon
#    probability of having profit less than 30000
prob=expon.cdf(30000,scale=1/scale)
print("probability is:",round(prob,2))
print("percentage is",round((prob*100),2))
```

```python
#    probability of having profit more than 60000
prob=1-expon.cdf(60000,scale=1/scale)
print("probability is:",round(prob,2))
print("percentage is",round((prob*100),2))
```

```python
#    probability of having profit between 40000 and 60000
prob=expon.cdf(60000,scale=1/scale)-expon.cdf(40000,scale=1/scale)
print("probability is:",round(prob,2))
print("percentage is",round((prob*100),2))
```
