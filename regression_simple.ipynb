```python
import numpy as np
```

```python
ybar=df.dist.mean()
```

```python
xbar=df.speed.mean()
```

```python
slope=np.sum((df.dist-ybar)*(df.speed-xbar))/np.sum((np.square(df.speed-xbar)))
```

```python
slope
```

```python
intercept=ybar-(xbar*slope)
```

```python
intercept
```

```python
df["ypred"]=intercept+(df.speed*slope)
```

```python
df
```

```python
plt.scatter(df.speed,df.dist)
plt.plot(df.speed,df.ypred,color="r")
```

```python
sns.kdeplot(df.ypred)
sns.kdeplot(df.dist)
```

```python
df["residual"]=df.dist-df.ypred
```

```python
sns.kdeplot(df.residual)
```

```python
#residual follows normal distribution
```
