```python
 import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters of the binomial distribution
n = 10  # Number of trials
p = 0.3  # Probability of success

# Number of samples and sample size
num_samples = 1000
sample_size = 100

# Generate the samples
samples = np.random.binomial(n, p, size=(num_samples, sample_size)) #it will result in 1000 samples and each sample it will have 100 trial

#randomly it generate binomial
# Calculate the sample means
sample_means = np.mean(samples, axis=1)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.title('Distribution of Sample Means (Binomial)')

# Show the plot
plt.show()
```
