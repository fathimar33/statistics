```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as ss
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df=pd.DataFrame(pd.read_csv("M6_T1_V4_BrakingDistance.csv"))
```

```python
formula="dist~speed"
```

```python
lm=smf.ols(formula,df).fit()
lm.summary()
```

```python
ypred=lm.predict()
```

```python
ypred
```

```python
df["ypred"]=ypred
```

```python
df
```

```python
df["residual"]=df["dist"]-df["ypred"]
```

```python
sns.kdeplot(df.residual) #residual follows normal distribution
```

```python
sns.scatterplot(df.speed,df.dist)
plt.plot(df.speed,df.ypred,color="r")
```

```python
sns.kdeplot(df.ypred)
sns.kdeplot(df.dist)
```
